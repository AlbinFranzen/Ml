# Prinicipal Components Analysis

PCA is a dimensionality reduction technique that aims to reduce the number of features in a dataset. First let's generate some data:

```python
# to make this notebook's output stable across runs
import numpy as np
np.random.seed(42)
```

```python
x = 2 * np.random.rand(100,1)
y = 4 + 3 * x + np.random.randn(100,1)
z = 3*x + y + np.random.randn(100,1)
```

```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot(111, projection='3d')


ax.scatter(x, y, z, c='b')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.show()
```

PCA essentially wants to take this data and project it onto a lower dimension while still preserving as much variance as possible. This can be done by finding the eigenvectors of the covariance matrix as implemented below:

```python
data = np.hstack((x,y))
covariance_matrix = 1/(len(data)-1)*(data-data.mean(0)).T.dot(data-data.mean(0))

eig_vals, eig_vecs = np.linalg.eig(covariance_matrix)
eig_vals
```

We see that the second eigenvalue is largest so we choose the second eigenvector:

```python
w = eig_vecs[:,1]
w = w.reshape((1,2))
```

```python
proj = data.dot(w.T)
```

```python
plt.scatter(proj,z,color="b",alpha=0.3)
plt.xlabel("1st principal component")
plt.ylabel("Z")
plt.show()
```

As we can see our dimensionality reduction of x and y maintained the features quite well in this case.
